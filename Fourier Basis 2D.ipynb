```python
#https://thepythoncodingbook.com/2021/08/30/2d-fourier-transform-in-python-and-fourier-synthesis-of-images/


#IMPORTANTE:o eixo de refer√™ncia para vizualiza√ß√£o dos √¢ngulos √© o y (vertical)

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)

#wavelength = 100 #comprimento de onda 'curto' (alta frequ√™ncia)
wavelength = 400 #comprimento de onda 'longo' (baixa frequ√™ncia)

#quanto menor o comprimento da onda, mais ondas cabem na mesma regi√£o do espa√ßo e,
#portanto, a frequ√™ncia da onda √© maior. Isso geralmente √© chamado de frequ√™ncia espacial.

#angle
angle = np.pi/20 # aprox. 10 graus
grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / wavelength
)
plt.set_cmap("viridis")
plt.imshow(grating)
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import numpy as np # Import numpy

# Re-create X, Y, and grating with compatible shapes
x = np.arange(-500, 501, 1)
X, Y = np.meshgrid(x, x)

wavelength = 400 # Use the same wavelength as before
angle = np.pi/20
grating = np.sin(
    2*np.pi*(X*np.cos(angle) + Y*np.sin(angle)) / wavelength
)

# Figure
fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(111, projection="3d")

# 3DAxes
ax.set_xlabel("X", size = 15)
ax.set_ylabel("Y", size = 15)
ax.set_zlabel("Vpp", size = 15)
ax.view_init(270,0)

#
ax.plot_surface(X, Y, grating, cmap = "viridis")

plt.show()
```
